```python
x = 1
y = 1
z = 1
solutions = 0

while z < 20:
    #print("{}² = {}² + {}² ? ".format(z, x, y))
    if z*z == x*x + y*y:
        print("on a trouvé une solution : ({},{},{})".format(x,y,z))
        solutions += 1
        x = 1
        y = 1
        z = z + 1
    else:
        if y < z-1:
            y = y + 1
        else:
            if x < z -1:
                x = x + 1
                y = x
            else:
                # On recommence pour le z suivant
                z = z + 1
                y = 1
                x = 1
                #print('z=',z)
   
print("Au total on a trouvé {} solution(s)".format(solutions))
    
```

```python
solutions = 0

for z in range(1,100):
    for y in range(1,z):
        for x in range(1,y):
            if x*x + y*y == z*z:  # ou x**2 (X puissance 2)
                print("on a trouvé une solution : ({},{},{})".format(x,y,z))
                solutions += 1
  
print("Au total on a trouvé {} solution(s)".format(solutions))

```
