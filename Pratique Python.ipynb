# Afficher la table de 5.


```python
# table de 5:
# 0 * 5 = 0
# 1 * 5 = 5
# 2 * 5 = 10 
...
# 12 * 5 = 60
```

```python
for table in range(0,21):
    print("Table de " + str(table))
    for i in range(0,13):
        print(str(i) + "*" + str(table) + " = " + str(i*table))


```
